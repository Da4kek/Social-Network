```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("embedding_data.csv")
data = data.drop("Unnamed: 0",axis=1)
data.head()
```

```python
data.corr()
```

## Cosine Similarity:

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_data = pd.DataFrame(cosine_similarity(data))
cosine_data.head()
```

```python
from sklearn.decomposition import PCA 
pcm = PCA(n_components=8)
prn = pcm.fit_transform(cosine_data)
prn_df = pd.DataFrame(prn,columns=['A','B',"C","D","E","F","G","H"])
prn_df.head()
```

```python
from sklearn.cluster import KMeans 
num_clusters = 8
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(cosine_data)
cosine_data['clusters'] = kmeans.predict(cosine_data)
cosine_data.head() 
```

```python
print(cosine_data['clusters'].unique())
```

```python
prn_df['clusters'] = cosine_data['clusters']
prn_df.clusters.value_counts()
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(1,1,1)

ax.set_xlabel("Circle1")
ax.set_ylabel("Circle2")
ax.set_title("Distinct Circles in the network")
cluster = cosine_data["clusters"].unique()
color = "b g r c m y k w".split(" ")
for cluster,color in zip(cluster,color):
    indexes = prn_df['clusters'] == cluster 
    ax.scatter(prn_df.loc[indexes,"A"],
               prn_df.loc[indexes,"B"],
               c = color,
               s =2)
ax.grid()
```

```python
cosine_data.to_csv("cosine_data.csv",index=False)
```
