```python
import pandas as pd 
import networkx as nx
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("../Compressed_data.csv")
data.head()
```

```python
network_1 = data['A'].values
network_2 = data['B'].values
```

```python
G = nx.Graph()
for i,j in zip(network_1,network_2):
    G.add_edge(int(i),int(j))
nx.draw(G,with_labels=True)
plt.show()
```
